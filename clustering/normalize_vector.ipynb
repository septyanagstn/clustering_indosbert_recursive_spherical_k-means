# Normalisasi L2 Vektor  

## Import Library

```python
import numpy as np
from sklearn.preprocessing import normalize
import os

print("Library berhasil diimpor.")
```

## Load Raw Vektor

```python
input_file = '../dataset/tickets_embedded.npy'

if os.path.exists(input_file):
    X = np.load(input_file)
    print(f"File {input_file} berhasil dimuat.")
    print(f"Dimensi vektor awal: {X.shape}")
    print(f"Magnitude rata-rata sebelum normalisasi: {np.mean(np.linalg.norm(X, axis=1)):.4f}")
else:
    print(f"Error: File {input_file} tidak ditemukan!")
```

## Normalisasi L2

### Melakukan Normalisasi L2 memastikan jumlah kuadrat dari elemen-elemen vektor adalah 1

```python
X_normalized = normalize(X, norm='l2')

print("Proses Normalisasi L2 selesai.")
print(f"Magnitude setelah normalisasi (harus 1.0): {np.linalg.norm(X_normalized[0]):.1f}")
```

## Save Vektor Hasil Normalisasi L2

```python
output_file = '../dataset/tickets_normalized.npy'

np.save(output_file, X_normalized)

print(f"Vektor yang telah dinormalisasi disimpan ke: {output_file}")
print("File ini siap digunakan sebagai input untuk Spherical K-Means.")
```
