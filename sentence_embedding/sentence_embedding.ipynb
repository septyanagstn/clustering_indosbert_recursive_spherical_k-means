## Import Library

```python
from sentence_transformers import SentenceTransformer
import pandas as pd
import numpy as np
import torch

print("Library berhasil diimpor.")
```

## Load Model IndoSBERT

```python
device = "cuda" if torch.cuda.is_available() else "cpu"

model_name = "denaya/indoSBERT-large"
print(f"Loading model {model_name} pada {device}...")

# Load model
model = SentenceTransformer(model_name, device=device)

print(f"Model {model_name} Large siap digunakan.")
```

## Load Data Preprocessed

```python
# Load data preprocessed
input_file = '../dataset/tickets_preprocessed.csv' 
df = pd.read_csv(input_file)

# preview 5 data teratas
print(f"Total data: {len(df)} baris")
display(df.head())


target_column = 'CLEAN_TEXT' 
df[target_column] = df[target_column].fillna("")

sentences = df[target_column].astype(str).tolist()
```

## Sentence Embedding

```python
print(f"Memulai proses embedding untuk {len(sentences)} kalimat...")

embeddings = model.encode(sentences)

print("Embedding selesai.")
print(f"Dimensi output: {embeddings.shape}") # Harusnya (n_data, 1024)

# Simpan hasil embedding agar tidak perlu mengulang proses ini
output_filename = '../dataset/tickets_embedded.npy'
np.save(output_filename, embeddings)
print(f"Vektor berhasil disimpan di: {output_filename}")
```
